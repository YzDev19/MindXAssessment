```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

#load the dataset
try:
    df = pd.read_csv('../data/mindx test dataset.csv')
except FileNotFoundError:
    df = pd.read_csv('data/mindx test dataset.csv')

print("Dataset Shape:", df.shape)
print(df.head())    

#check missing values
print("\nMissing Values:")
print(df.isnull().sum())

#analysis to confirm fuel consumption is related to  distance & ship type
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='distance', y='fuel_consumption', hue='ship_type')
plt.title('Fuel Consumption vs distance by Ship Type')
plt.show()

```
